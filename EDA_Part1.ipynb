```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

##Display all the columns of the dataframes
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('train.csv')
```

```python
dataset.head()
```

```python
##print the shape of the dataset with rows and columns
print(dataset.shape)
```

```python
#Missing values
dataset.isnull().sum()
```

```python
features_with_NaN=[features for features in dataset.columns if dataset[features].isnull().sum()>=1]
```

```python
#these are the features_with_NaN values
features_with_NaN
```

```python
#dataset['Electrical'].isnull().mean()
for features in features_with_NaN:
    print(features,np.round(dataset[features].isnull().mean()*100,4),'% missing values')
```

```python

for features in features_with_NaN:
    data=dataset.copy()
    # let' s make a variable that indicate 1 if the observation was missing or 0 otherwise
    data[features]=np.where(data[features].isnull(),1,0)
    # let's calculate the mean Sales price where the information is missing or present
    data.groupby(features)['SalePrice'].median().plot.bar()
    plt.title(features)
    plt.show()
```

Here with the relation between the missing values and the dependent variable is clearly visible.So we need to replace these nan values with something meaningful which we will do in the Feature Engineering section

```python
##list of numerical variable
dataset['SalePrice'].dtype!='O'
```

```python
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtype!='O']
```

```python
print(len(numerical_features))
dataset[numerical_features].head()
```

```python
## Temporal variables(Eg:DateTime variables)
year_feature=[feature for feature in numerical_features if 'Yr'in feature or 'Year' in feature]
```

```python
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs Year Sold')
```

```python
#here we will compare the difference between all years features with saleprice
data=dataset.copy()
for feature in year_feature:
    if feature!='YrSold':
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

within all the yearwise information. it is clearly visible that the new houses are more costier that the older one.looks like a pareto power low distribution.


```python
### numerical variable - 2Types
## Continuous variable(can be anything) and Discrete variable(whole number)
Discrite_features=[feature for feature in numerical_features if len(dataset[feature].unique())<=25]
```

```python
print(len(Discrite_features))
```

```python
dataset[Discrite_features].head()
```

```python
data=dataset.copy()
for feature in Discrite_features:
    data.groupby(feature)['SalePrice'].median().plot().bar
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
    
```

```python
#There is a relationship between Discreate feature and sale price
```

```python
continuous_feature=[feature for feature in numerical_features if feature not in Discrite_features+year_feature]
```

```python
print(len(continuous_feature))
```

```python
### lets analysis the continuous values by creating histogram to understand the observation
data=dataset.copy()
for feature in continuous_feature:
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```
